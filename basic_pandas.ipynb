```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/panel_data.csv')
df.head()
```

```python
df['brand'].unique()
```

```python
df[df.brand.str.contains('BMW|VOLVO|KIA')].nlargest(5, 'total')
```

```python
t= df.drop('month', axis=1).sum(numeric_only=True)
t
```

```python
t.sort_values().plot(kind='barh')
```

```python
df.drop('month', axis=1).groupby('brand').sum().nlargest(5, 'total').plot(kind='barh',subplots=True,layout=(2,4), figsize=(12,5),sharey=True)
```
